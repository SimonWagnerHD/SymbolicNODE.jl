```julia
using DelimitedFiles, Plots
```

```julia
node_train = readdlm("../data/node_train100.csv", ',')
node_valid = readdlm("../data/node_valid100.csv", ',')
sindy_node_train = readdlm("../data/sindy_node_train100.csv", ',')
sindy_node_valid = readdlm("../data/sindy_node_valid100.csv", ',')
sindy_node_add_train = readdlm("../data/sindy_node_add_train100.csv", ',')
sindy_node_add_valid = readdlm("../data/sindy_node_add_valid100.csv", ',')
sindy_node_cat_train = readdlm("../data/sindy_node_cat_train100.csv", ',')
sindy_node_cat_valid = readdlm("../data/sindy_node_cat_valid100.csv", ',');
```

```julia
plot1 = plot(node_train, label="node")
plot!(sindy_node_train, label="sindy_node")
#plot!(sindy_node_add_train, label="sindy_node_add")
plot!(sindy_node_cat_train, label="sindy_node_cat")

plot2 = plot(node_valid, label="node")
plot!(sindy_node_valid, label="sindy_node")
#plot!(sindy_node_add_valid, label="sindy_node_add")
plot!(sindy_node_cat_valid, label="sindy_node_cat")

plot(plot1, plot2, layout = grid(1, 2), size=(1000,400))
```
